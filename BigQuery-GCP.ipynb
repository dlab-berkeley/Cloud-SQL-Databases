- **Wikipedia Pageviews**: `bigquery-public-data.wikipedia.pageviews_2015`

```
SELECT *
FROM `bigquery-public-data.wikipedia.pageviews_2015`
LIMIT 10;
```

### **Query Error?**

### **![][image4]**

The error you're encountering is related to the fact that the table you're trying to query (`bigquery-public-data.wikipedia.pageviews_2015`) is a partitioned table. In BigQuery, partitioned tables require a filter on the partition column to limit the number of partitions scanned and thus improve query efficiency.

For the `wikipedia.pageviews_2015` table, the partitioning column is `datehour`. Therefore, you need to include a condition to filter on this `datehour` column. Here's how you can modify your query to resolve the error:

### **Solution**

Add a filter on the `datehour` column to limit the partitions being queried. For example, you can limit the query to a specific date or date range:

```
SELECT *
FROM `bigquery-public-data.wikipedia.pageviews_2015`
WHERE datehour BETWEEN '2015-01-01 00:00:00' AND '2015-01-01 23:00:00'
LIMIT 10;
```

### **Explanation**:

- The `datehour` column stores the date and hour information for each partition.  
- By adding a filter condition, you ensure that only a specific partition or range of partitions is scanned, which improves query performance and complies with BigQuery's partitioned table rules.

### **Examples of Filters on the `datehour` Column**:

1. **Range of Hours**:  
     
   You can also specify a range, for example between October and December 2016:

```
SELECT *
FROM `bigquery-public-data.wikipedia.pageviews_2016`
WHERE datehour BETWEEN '2016-10-01 00:00:00' AND '2016-12-01 00:00:00'
LIMIT 10;
```

### **Tips**:

- Always include a filter on the partition column (`datehour` in this case) when querying partitioned tables.  
- You can use a wide date range if you want more data, but ensure that the range is feasible given the data and the resource quotas of your project.

### **Further Information**:

Partitioned tables in BigQuery allow for better management of large datasets by dividing them into smaller, more manageable segments called partitions. Adding a filter on the partition column helps BigQuery "prune" the table, only scanning the relevant partitions and skipping the rest, resulting in reduced query costs and better performance.

[Introduction to partitioned tables | BigQuery | Google Cloud](https://cloud.google.com/bigquery/docs/partitioned-tables)  
[https://cloud.google.com/bigquery/docs/partitioned-tables](https://cloud.google.com/bigquery/docs/partitioned-tables)  
A partitioned table is divided into segments, called partitions, that make it easier to manage and query your data. By dividing a large table into smaller partitions, you can improve query performance and control costs by reducing the number of bytes read by a query. You partition tables by specifying a partition column which is used to segment the table.

If a query uses a qualifying filter on the value of the partitioning column, BigQuery can scan the partitions that match the filter and skip the remaining partitions. This process is called pruning.

In a partitioned table, data is stored in physical blocks, each of which holds one partition of data. Each partitioned table maintains various metadata about the sort properties across all operations that modify it. The metadata lets BigQuery more accurately estimate a query cost before the query is run.
